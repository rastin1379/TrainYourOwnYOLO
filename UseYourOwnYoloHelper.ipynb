```python
file_list = os.listdir("./Highway_Dataset/Labels_Train")
file_list.remove('.DS_Store')
print(file_list)
```

```python
data = [] # pd.concat takes a list of dataframes as an agrument
for textfile in file_list:
    frame = pd.read_csv("./Highway_Dataset/Labels_Train/"+textfile, sep=" ",header=None)
    frame['filename'] = os.path.basename(textfile).replace("txt","jpg")
    data.append(frame)
```

```python
data
```

```python
yoloframe = pd.concat(data, ignore_index=True) #dont want pandas to try an align row indexes
```

```python
yoloframe.head(10)
```

```python
# yoloframe.rename(columns={"0": "object-class", "1": "x","2": "y","3": "width","4": "height","filename":"Image"}, inplace=True)
yoloframe.columns = ['object-class', 'x_center', 'y_center', 'width', 'height','image']

yoloframe.head(20)

# yoloframe.to_csv("Annotations-export.csv", index=False)
# frame = pd.read_csv("Train/Labels/"+file_list[0], sep=" ", header=None)
# frame
```

```python
useown_dataframe = pd.DataFrame(columns=['image','xmin','ymin','xmax','ymax','label'])
```

```python
useown_dataframe
```

```python
useown_dataframe["image"]=yoloframe["image"]
useown_dataframe["xmin"]=yoloframe["x_center"] - yoloframe["width"] / 2
useown_dataframe["ymin"]=yoloframe["y_center"] - yoloframe["height"] / 2
useown_dataframe["label"]=yoloframe["object-class"]
useown_dataframe["xmax"]=yoloframe["x_center"] + yoloframe["width"] / 2
useown_dataframe["ymax"]=yoloframe["y_center"] + yoloframe["height"] / 2
```

```python
useown_dataframe
```

```python
# maping labels to names
objects = ["Truck", "Bus", "Car"]
useown_dataframe["label"] = useown_dataframe["label"].apply(lambda x: objects[x])
```

```python
image_list = os.listdir("./Highway_Dataset/Train_Data")
image_list.remove('.DS_Store')
image_list
```

```python
# scaling x & y
for imagefile in image_list:
    im = Image.open("./Highway_Dataset/Train_Data/" + imagefile)
    width, height = im.size
    useown_dataframe.loc[useown_dataframe.image == imagefile, 'ymin'] *= height
    useown_dataframe.loc[useown_dataframe.image == imagefile, 'ymax'] *= height
    useown_dataframe.loc[useown_dataframe.image == imagefile, 'xmin'] *= width
    useown_dataframe.loc[useown_dataframe.image == imagefile, 'xmax'] *= width
```

```python
useown_dataframe
```

```python
useown_dataframe.to_csv("Annotations-export.csv", index=False)
```
